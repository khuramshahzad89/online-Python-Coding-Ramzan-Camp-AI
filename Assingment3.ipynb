```python
# Q1. Make a calculator using Python with addition, subtraction, multiplication, division and power.
# Answer Q1
def calculator(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "divide":
        return a / b if b != 0 else "Division by zero not allowed"
    elif operation == "power":
        return a ** b
    else:
        return "Invalid operation"


print("Addition:", calculator(10, 5, "add"))
print("Subtraction:", calculator(10, 5, "subtract"))
print("Multiplication:", calculator(10, 5, "multiply"))
print("Division:", calculator(10, 5, "divide"))
print("Power:", calculator(2, 3, "power"))
```

```python
# Q2. Write a program to check if there is any numeric value in list using for loop.
# Answer Q2:

my_list = ["mango", "banana", 5, "stowbery", 7.5, "apricote"]
found_numeric = False

for item in my_list:
    if isinstance(item, (int, float)):
        print("Numeric value found:", item)
        found_numeric = True
if not found_numeric:
    print("No numeric values found in the list.")

```

```python
# Q3. Write a Python script to add a key to a dictionary.
# Answer Q3:

my_dict = {"name": "Ali", "age": 25}
print("Original Dictionary:", my_dict)

my_dict["city"] = "Karachi"  # Adding new key
print("Updated Dictionary:", my_dict)
```

```python
# Q4. Write a Python program to sum all the numeric items in a dictionary.
# Answer Q4:
data = {"a": 10, "b": "hello", "c": 20, "d": 5.5}
total = sum(value for value in data.values() if isinstance(value, (int, float)))

print("Sum of numeric items:", total)
```

```python
# Q5. Write a program to identify duplicate values from list. 
# Answer Q5:
numbers = [1, 2, 3, 2, 4, 5, 1, 6]
duplicates = set([x for x in numbers if numbers.count(x) > 1])

print("Duplicate values:", duplicates)
```

```python
# Q6. Write a Python script to check if a given key already exists in a dictionary
# Answer Q6:
student = {"name": "Sara", "age": 22, "grade": "A"}

key_to_check = "age"
if key_to_check in student:
    print(f"Key '{key_to_check}' exists in dictionary.")
else:
    print(f"Key '{key_to_check}' does not exist.")
```
